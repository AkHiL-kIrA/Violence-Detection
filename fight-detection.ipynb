### Fight Training Dataset

```python
import os
import cv2


# Input and output folder paths
video_folder = "./Peliculas/train/fights/"    
frame_folder = './myData/train/fights/'

# Create the frame folder if it doesn't exist
os.makedirs(frame_folder, exist_ok=True)

# Function to extract frames from a video and save them to the frame folder
def extract_frames(video_path, output_folder):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Check if the video opened successfully
    if not cap.isOpened():
        print(f"Error opening video file: {video_path}")
        return

    # Variables for frame count and interval
    frame_count = 0
    frame_interval = 10  # Extract every 30 frames, for example

    # Loop through the video frames
    while cap.isOpened():
        ret, frame = cap.read()

        # Check if the frame was read successfully
        if not ret:
            break

        # Extract frames at the specified interval
        if frame_count % frame_interval == 0:
            # Save the frame as an image in the output folder
            frame_filename = os.path.join(output_folder, f'{os.path.basename(video_path)}frame{frame_count}.jpg')
            cv2.imwrite(frame_filename, frame)
            print(f"Saved frame {frame_count} from {video_path} as {frame_filename}")

        frame_count += 1

    # Release the video capture object
    cap.release()

# Iterate over each video file in the video folder
for root, dirs, files in os.walk(video_folder):
    for file in files:
        if file.endswith((".mp4",".avi",'.mov','.mkv','.mpg')):  # Check if the file is a video file
            video_path = os.path.join(root, file)
            extract_frames(video_path, frame_folder)
```

```python
import os
import cv2


# Input and output folder paths
video_folder = "./Peliculas/val/notfights/"    
frame_folder = './myData/val/notfights/'

# Create the frame folder if it doesn't exist
os.makedirs(frame_folder, exist_ok=True)

# Function to extract frames from a video and save them to the frame folder
def extract_frames(video_path, output_folder):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Check if the video opened successfully
    if not cap.isOpened():
        print(f"Error opening video file: {video_path}")
        return

    # Variables for frame count and interval
    frame_count = 0
    frame_interval = 10  # Extract every 30 frames, for example

    # Loop through the video frames
    while cap.isOpened():
        ret, frame = cap.read()

        # Check if the frame was read successfully
        if not ret:
            break

        # Extract frames at the specified interval
        if frame_count % frame_interval == 0:
            # Save the frame as an image in the output folder
            frame_filename = os.path.join(output_folder, f'{os.path.basename(video_path)}frame{frame_count}.jpg')
            cv2.imwrite(frame_filename, frame)
            print(f"Saved frame {frame_count} from {video_path} as {frame_filename}")

        frame_count += 1

    # Release the video capture object
    cap.release()

# Iterate over each video file in the video folder
for root, dirs, files in os.walk(video_folder):
    for file in files:
        if file.endswith((".mp4",".avi",'.mov','.mkv','.mpg')):  # Check if the file is a video file
            video_path = os.path.join(root, file)
            extract_frames(video_path, frame_folder)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1/255)
train_generator = train_datagen.flow_from_directory(directory="./myData/train/",
                                                        target_size=(300, 300),
                                                        batch_size=128,
                                                        class_mode="binary")
```

```python
val_datagen = ImageDataGenerator(rescale=1/255)
val_generator = val_datagen.flow_from_directory(directory="./myData/val/",batch_size=128,class_mode="binary",target_size=(300,300))
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 300x300 with 3 bytes color
    # This is the first convolution
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(300, 300, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    # The second convolution
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The third convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The fourth convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The fifth convolution
    tf.keras.layers.Conv2D(256, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),
    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'),
    # Only 1 output neuron. It will contain a value from 0-1 where 0 for 1 class ('horses') and 1 for the other ('humans')
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.save("Violence.h5")
```

```python
from tensorflow.keras.optimizers import RMSprop

model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(learning_rate=0.001),
              metrics=['accuracy'])
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    '''
    Halts the training when the loss falls below 0.4

    Args:
      epoch (integer) - index of epoch (required but unused in the function definition below)
      logs (dict) - metric results from the training epoch
    '''

    # Check the loss
    if(logs.get('accuracy') > 0.98):

      # Stop if threshold is met
      print("\naccuracy is greater than 0.98 so cancelling training!")
      self.model.stop_training = True

# Instantiate class
callbacks = myCallback()
```

```python
history = model.fit(
      train_generator,
      steps_per_epoch=8,
      epochs=15,
      verbose=1,
      validation_data = val_generator,
      validation_steps=8, callbacks= [callbacks])
```
